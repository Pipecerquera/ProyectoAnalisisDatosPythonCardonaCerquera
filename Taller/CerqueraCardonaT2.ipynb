### 2. Crear el Dataset
Ejecuta la siguiente celda para crear los DataFrames que usaremos en los ejercicios.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo
sns.set_theme(style='whitegrid')
np.random.seed(42)

# Dataset 1: Ventas mensuales
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
ventas_2022 = [45, 52, 48, 65, 70, 85, 90, 88, 75, 80, 95, 110]
ventas_2023 = [50, 58, 55, 72, 78, 92, 98, 95, 82, 88, 105, 120]

df_ventas = pd.DataFrame({
    'Mes': meses,
    'Ventas_2022': ventas_2022,
    'Ventas_2023': ventas_2023
})

# Dataset 2: Datos de empleados
n_empleados = 150
df_empleados = pd.DataFrame({
    'Edad': np.random.normal(35, 8, n_empleados).astype(int).clip(22, 65),
    'Salario': np.random.normal(50000, 15000, n_empleados).clip(25000, 120000),
    'Años_Experiencia': np.random.normal(8, 4, n_empleados).clip(0, 30),
    'Departamento': np.random.choice(['Ventas', 'IT', 'Marketing', 'RRHH'], n_empleados),
    'Satisfaccion': np.random.randint(1, 11, n_empleados)
})

# Dataset 3: Productos
productos = ['Laptop', 'Mouse', 'Teclado', 'Monitor', 'Auriculares']
unidades_vendidas = [120, 450, 380, 210, 290]

df_productos = pd.DataFrame({
    'Producto': productos,
    'Unidades': unidades_vendidas
})

print('Datasets creados exitosamente')
print(f'\ndf_ventas: {df_ventas.shape}')
print(f'df_empleados: {df_empleados.shape}')
print(f'df_productos: {df_productos.shape}')
```

### 3. Gráfico de Líneas Simple

### 15. Heatmap de Correlación
Crea un heatmap que muestre la matriz de correlación de las variables numéricas de `df_empleados`. Usa `sns.heatmap()` con `annot=True`.

```python
# Tu código aquí
corr = df_empleados[['Edad', 'Salario', 'Años_Experiencia', 'Satisfaccion']].corr()
plt.figure(figsize=(8, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación - Empleados')
plt.show()
```

```python
corr
```

El heatmap presentado revela la matriz de correlación entre variables numéricas como edad, salario, años de experiencia y nivel de satisfacción de los empleados, utilizando el método de Spearman. Los valores cercanos a cero indican que no existe una relación significativa entre estas variables, lo que sugiere que, por ejemplo, tener más años de experiencia no garantiza un mayor salario ni una mayor satisfacción laboral. Esta falta de correlación puede reflejar una estructura organizacional donde los factores como edad o experiencia no influyen directamente en la compensación o bienestar, lo cual podría ser relevante para revisar políticas internas de desarrollo profesional y motivación.

### 17. Subplots (Múltiples Gráficos)
Crea una figura con 2 subplots (1 fila, 2 columnas):
- Subplot 1: Histograma de edades
- Subplot 2: Histograma de salarios

Usa `plt.subplots()`.

```python
# Tu código aquí
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].hist(df_empleados['Edad'], bins=15)
axes[0, 0].set_title('Distribución de Edades')
axes[0, 0].set_xlabel('Edad')
axes[0, 0].set_ylabel('Frecuencia')

axes[0, 1].hist(df_empleados['Salario'], bins=15, color='orange')
axes[0, 1].set_title('Distribución de Salarios')
axes[0, 1].set_xlabel('Salario')
axes[0, 1].set_ylabel('Frecuencia')

axes[1, 0].hist(df_empleados['Edad'], bins=15)
axes[1, 0].set_title('Distribución de Edades')
axes[1, 0].set_xlabel('Edad')
axes[1, 0].set_ylabel('Frecuencia')

axes[1, 1].hist(df_empleados['Salario'], bins=15, color='orange')
axes[1, 1].set_title('Distribución de Salarios')
axes[1, 1].set_xlabel('Salario')
axes[1, 1].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

Este conjunto de histogramas muestra la distribución de edades y salarios de los empleados en una cuadrícula de 2x2, facilitando la comparación visual. Los gráficos revelan cómo se concentran los valores en ciertos rangos: por ejemplo, las edades parecen agruparse en torno a una franja específica, lo que podría indicar una plantilla con predominancia de cierta generación. En cuanto a los salarios, se observa una dispersión que podría reflejar diferencias en cargos, experiencia o políticas de compensación. Aunque los gráficos están duplicados, esta estructura permite observar patrones repetidos y verificar consistencia en los datos. Es una herramienta útil para detectar sesgos o tendencias en la composición del personal.

### 19. Personalización Avanzada
Crea un scatter plot de `Años_Experiencia` vs `Salario` con las siguientes personalizaciones:
- Tamaño de figura: 10x6
- Color de puntos: verde
- Transparencia (alpha): 0.6
- Tamaño de puntos: 50
- Grid activado
- Título con tamaño de fuente 16
- Etiquetas de ejes con tamaño de fuente 12

```python
# Tu código aquí
plt.figure(figsize=(10, 6))
plt.scatter(df_empleados['Años_Experiencia'], df_empleados['Salario'],
            color='green', alpha=0.6, s=50)
plt.title('Relación entre Años de Experiencia y Salario', fontsize=16)
plt.xlabel('Años de Experiencia', fontsize=12)
plt.ylabel('Salario', fontsize=12)
plt.grid(True)
plt.show()
```

Este gráfico de dispersión muestra cómo se relacionan los años de experiencia con el salario de los empleados. Aunque se podría esperar una tendencia ascendente —es decir, que más experiencia conlleve mayor salario— los puntos verdes están bastante dispersos, lo que indica una baja correlación entre ambas variables. Esto sugiere que factores adicionales, como el departamento, el nivel educativo o el cargo específico, podrían estar influyendo más en la remuneración que la experiencia por sí sola. La visualización es útil para detectar inconsistencias en políticas salariales y evaluar si la experiencia está siendo valorada de forma equitativa dentro de la organización.
